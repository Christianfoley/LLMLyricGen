```python
import pandas as pd
import numpy as np
from datasets import load_dataset
```

```python
lyric_data = load_dataset("cs182project/182-final-project")
lyric_data
```

```python
test = lyric_data['test']
```

```python
test_df = pd.DataFrame(test)
test_df
```

```python
def extact_prompt(s):
    return [s[0]['value']]
```

```python
test_df['turns'] = test_df['conversations'].map(extact_prompt)
test_df
```

```python
test_df['category'] = ['music'] * len(test_df)
test_df
```

```python
test_df.rename({'id': 'question_id'}, axis=1, inplace=True)
test_df
```

```python
test_df[['question_id', 'category', 'turns']].to_json('fastchat/llm_judge/data/lyric/question.jsonl', orient='records', lines=True)
```
