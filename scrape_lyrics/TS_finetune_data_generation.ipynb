```python
import pandas as pd
import requests
import json
```

```python
def clean_text(text):
    return text.split("(")[0].strip().lower().replace("\u200b","")
def clean_list(l):
    return [clean_text(x) for x in l]
```

```python
training = pd.read_csv("https://raw.githubusercontent.com/Christianfoley/LLMLyricGen/main/SongLyricsScraper/1000_songs_no_annotation.csv")
test = pd.read_csv("https://raw.githubusercontent.com/Christianfoley/LLMLyricGen/main/SongLyricsScraper/modern_songs.csv")
```

```python
training = training[training["artist"] == "Taylor Swift"]
test = test[test["artist"] == "Taylor Swift"]
```

```python
taylor_songs = list(pd.read_csv("https://raw.githubusercontent.com/shaynak/taylor-swift-lyrics/main/songs.csv")["Title"])
```

```python
taylor_songs = [clean_text(x) for x in taylor_songs]
```

```python
all_songs = pd.read_csv("./final_cleaned.csv")
taylor_all = all_songs[all_songs["artist"] == "Taylor Swift"]
taylor_all
```

```python
unincluded = ['Gracie','You Do','Welcome Distraction','This Here Guitar','This Is Really Happening','Tell Me','Stupid Boy','Only The Young','Permanent Marker','Need','Need You Now','Her','I’d Lie','Kid in the Crowd','Love to Lose','Lucky You','All Night Diner','American Boy','Angelina','Birch','Both of Us','Brought Up That Way','Can I Go with You','Check Out This View','Cross My Heart','Drama Queen']
```

```python
cleaned_taylor_all = taylor_all[taylor_all["song"].apply(lambda x: clean_text(x) in taylor_songs or x in unincluded)]
```

```python
available_taylor_ids = list(cleaned_taylor_all["id"])
available_taylor_song = list(cleaned_taylor_all["song"])
```

```python
unavailable_ids = list(training["id"])
unavailable_songs = list(training["song"])
unavailable_songs = [clean_text(x) for x in unavailable_songs]
```

```python
cleaned_taylor_all["song_stripped"] = cleaned_taylor_all["song"].apply(clean_text)
```

```python
cleaned_taylor_all = cleaned_taylor_all.drop_duplicates(subset="song_stripped")
```
