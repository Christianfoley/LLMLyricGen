```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import sys
sys.path.insert(0, "../training_curves")
```

```python
lr_data = pd.read_csv("lr_search.csv")[['atomic-surf-86 - train/loss', 'bumbling-lion-85 - train/loss', 'spring-frog-84 - train/loss']]
lr_data.rename({'atomic-surf-86 - train/loss': 'lr = 2e-6', 'bumbling-lion-85 - train/loss': 'lr = 2e-4', 'spring-frog-84 - train/loss': 'lr = 2e-5'}, axis=1, inplace=True)
lr_data
```

```python
sns.set_style("darkgrid")
sns.lineplot(lr_data, palette=sns.color_palette("pastel"))
plt.title('Full Fine-tuning Traing Loss for different Learning Rates')
plt.xlabel('Step')
plt.ylabel('Cross Entropy Loss')
```

```python
base_data = pd.read_csv("diff_base_models.csv")[['peachy-cosmos-95 - train/loss', 'sage-river-94 - train/loss']]
base_data.rename({'peachy-cosmos-95 - train/loss': 'Base Model = Llama-2-7B-chat', 'sage-river-94 - train/loss': 'Base Model = Llama-2-7B'}, axis=1, inplace=True)
```

```python
sns.lineplot(base_data, palette=sns.color_palette("pastel"))
plt.title("Training Curves for different Base Models")
plt.xlabel("Step")
plt.ylabel("Cross Entropy Loss")
```

```python
lora_data = pd.read_csv("lora_rank_search.csv")[['pious-hill-96 - train/loss', 'atomic-plasma-92 - train/loss', 'swept-serenity-90 - train/loss']]
lora_data.rename({'pious-hill-96 - train/loss': 'Lora Rank = 256', 'atomic-plasma-92 - train/loss': 'Lora Rank = 16', 'swept-serenity-90 - train/loss': 'Lora Rank = 8'}, axis=1, inplace=True)
lora_data.dropna(inplace=True)
```

```python
sns.lineplot(lora_data, palette=sns.color_palette("pastel"))
plt.title('LoRA Fine-tuning loss for different Lora Ranks')
plt.xlabel('Step')
plt.ylabel('Cross Entropy Loss')
```

```python
taylor_data = pd.read_csv("taylor_swift_loss.csv")[['Step', 'rare-sound-100 - train/loss', 'efficient-universe-99 - train/loss']]
taylor_data.rename({'rare-sound-100 - train/loss' : 'Base Model: Llama-7B-Chat', 'efficient-universe-99 - train/loss': 'Lyre-Chat-Checkpoint-100'})
taylor_data.set_index('Step', inplace=True, drop=True)
taylor_data
```
